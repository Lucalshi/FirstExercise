```python
import pm4py
from pm4py.objects.log.importer.xes import importer as xes_importer

log = xes_importer.apply("../data/BPIC2017.xes.gz")
```

```python
bpmn_graph = pm4py.read_bpmn("FirstExercise BPMN Model-4.bpmn")
```

```python
net, im, fm = pm4py.convert_to_petri_net(bpmn_graph)
```

### Fitness

```python
fitness = pm4py.fitness_token_based_replay(log, net, im, fm)
fitness
```

### Precision

```python
from pm4py.algo.evaluation.precision import algorithm as precision_evaluator

precision = precision_evaluator.apply(log, net, im, fm, variant=precision_evaluator.Variants.ETCONFORMANCE_TOKEN)
precision
```

### Generalization

```python
from pm4py.algo.evaluation.generalization import algorithm as generalization_evaluator

generalization   = generalization_evaluator.apply(log, net,  im, fm)
generalization
```

### Simplicity

```python
from pm4py.algo.evaluation.simplicity import algorithm as simplicity_evaluator

simplicity   = simplicity_evaluator.apply(net)
simplicity
```
